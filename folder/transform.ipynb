```python
import pandas as pd

csv = pd.read_csv('data_rules.csv')
```

```python
csv[['Unit_InfoObject_Source',
     'Assignment_InfoObject_Source']] = csv['Source'].str.extract(r'Unit InfoObject\s:\s(.*)\s*Assignment InfoObject\s:\s*(.*)')
csv[['Unit_InfoObject_Target',
     'ConvExit_Target',
     'Aggregatn_Target']] = csv['Target'].str.extract(r'Unit InfoObject\s:\s(.*)\s*ConvExit\s:\s(.*)\s*Aggregatn\s:\s*(.*)')
csv[['Referential_Integrity_Target',
     'Set_Conversion_Exist_Target']] = csv['Target'].str.extract(r'Referential Integrity\s:\s*(.*)\s*Set Conversion Exit:\s*(.*)')
```

```python
csv.head(5)
```

```python
csv.drop(['Target','Source'], axis=1, inplace=True)
```

```python
csv.to_csv('data_rules_transformed.csv', encoding="utf-8", index=False)
```
